```python
import numpy as np
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

weather = pd.read_csv('weatherHistory-mini.csv')
print(weather.isna().sum()) # Find a data is lose in your file
```

```python
boolean_missing = weather.isna().any(axis=1)  # Show all row that are missing 
weather_missing = weather[boolean_missing]
print(weather_missing)
```

```python
missing_vars = ['Wind Speed (km/h)'] #Set value of Wind Speed (km/h) from missing values back to simple data with none missing values
indpend_vars = ['Temperature (C)','Wind Bearing (degrees)']
imputer = IterativeImputer()
imputed_data = imputer.fit_transform(weather[indpend_vars+missing_vars])
print(imputed_data)
```

```python
#Take Wind Speed where is non missing values from inputed_data 
# take to new DataFrame name weather and display it
weather[missing_vars] = imputed_data[:,-len(missing_vars):] #or weather[missing_vars] = imputed_data[:-,1:]
display(weather)
```
