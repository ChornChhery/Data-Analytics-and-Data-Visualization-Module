```python
import matplotlib.pyplot as pltimport 
import seaborn as sns
import matplotlib.pyplot as plt
# Sample data
data = {
    'sex': [0, 1, 0, 1, 0],
    'mobidity': [1, 2, 0, 2, 1]
}

dt1 = pd.DataFrame(data)

# Convert 'sex' column to categorical and rename levels
dt1['sex'] = dt1['sex'].astype('category')

# Check the existing categories
print("Existing 'sex' categories:", dt1['sex'].cat.categories)

# Rename categories
dt1['sex'] = dt1['sex'].cat.rename_categories(["Male", "Female"])

# Convert 'mobidity' column to categorical
dt1['mobidity'] = dt1['mobidity'].astype('category')

# Check the existing categories
print("Existing 'mobidity' categories:", dt1['mobidity'].cat.categories)

# Rename categories
dt1['mobidity'] = dt1['mobidity'].cat.rename_categories(["Mild", "Severe", "No Disease"])

# Display the DataFrame
print(dt1)

# Plotting the distributions using bar charts
plt.figure(figsize=(10, 5))

# Bar plot for 'sex'
plt.subplot(1, 2, 1)
sns.countplot(x='sex', data=df1)
plt.title('Distribution of Sex')

# Bar plot for 'mobidity'
plt.subplot(1, 2, 2)
sns.countplot(x='mobidity', data=df1)
plt.title('Distribution of Mobidity')

plt.tight_layout()
plt.show()

```
